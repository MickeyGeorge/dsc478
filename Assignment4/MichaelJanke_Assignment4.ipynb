```python
import numpy as np
import pylab as pl
import pandas as pd
from sklearn.cluster import KMeans 
from sklearn import preprocessing

np.set_printoptions(precision=2, linewidth=500, suppress=True, threshold=np.nan)

from IPython.display import display
pd.options.display.max_columns = None
pd.options.display.max_rows = None
```

```python
## 1.a load data

## Rows = images
## Columns = features

data = pd.read_csv('segmentation_data.txt', header=None)
classes = pd.read_table('segmentation_classes.txt', sep='\t', header=None)
```

```python
data.head(10)
```

```python
classes.head(10)
```

```python
## 1.a Perform Min-Max Normalization

min_max_scaler = preprocessing.MinMaxScaler().fit(data)
data_norm = min_max_scaler.transform(data)

data_norm[0:5]
```

```python
## 1.b perfrom Kmeans clustering

kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
kmeans.fit(data_norm)
```

```python
clusters = kmeans.predict(data_norm)
```

```python
clusters
```

```python
centroids = kmeans.cluster_centers_
```

```python
# 1.b print centroids

centroids
```

```python
# 1.b completeness and homogeneity

from sklearn.metrics import completeness_score, homogeneity_score
```

```python
print completeness_score(classes[1],clusters)
```

```python
print homogeneity_score(classes[1],clusters)
```

```python
# 1.c Perform PCA

from sklearn import decomposition
pca = decomposition.PCA(n_components=6)

data_trans = pca.fit(data_norm).transform(data_norm)
print data_trans
```

```python
# six components explain 95% of the variance

print(pca.explained_variance_ratio_)
```

```python
# 1.c Perform kmeans on reduced dimensions

kmeans = KMeans(n_clusters=7, max_iter=500, verbose=1) # initialization
kmeans.fit(data_trans)
```

```python
clusters_pca = kmeans.predict(data_trans)
clusters_pca
```

```python
centroids_pca = kmeans.cluster_centers_
centroids_pca
```

```python
print completeness_score(classes[1],clusters_pca)
```

```python
print homogeneity_score(classes[1],clusters_pca)
```

### 1.e discuss

#### Scores for full data

Completeness Score = 0.6116744999910891 <br>
Homogeneity Score = 0.6099656393147241 <br>

#### Scores after PCA

Completeness Score = 0.6108391577738163 <br>
Homogeneity Score = 0.6091689230222487 <br>

The scores before and after PCA are essentially the same. This illustrates that the performance of a model can remain unchanged when utilizing PCA to reduce dimensionality. PCA only retains those features that provide value to the model.
