## 3. z-score normalization

```python
## 3. z-score normalization of income values

income_z = (bank_df["income"] - bank_df["income"].mean()) / bank_df["income"].std()
income_z.head(5)
```

## 4. Discretize age attribute

```python
## 4. Discretize age

age_bins = pd.qcut(bank_df.age, [0, .33, .66, 1], labels=["young", "mid-age", "old"], retbins=True)
age_bins
```

## 5. Min-Max Normalization

```python
## 5. Min-Max Normalization

## The following results in an error
##
## bank_norm = (bank_bin - bank_bin.min()) / (bank_bin.max()-bank_bin.min())
## bank_norm.head(10)
##
## ValueError: operands could not be broadcast together with shapes (4800,) (8,)
##
## Therefore I explicitly applied to the numeric columns

bank_norm=(bank_df[["income", "age", "children"]]-bank_df[["income", "age", "children"]].min())/(bank_df[["income", "age", "children"]].max()-bank_df[["income", "age", "children"]].min())
bank_norm.head(10)
```

## 6. Convert table to spreadsheet

```python
## 6.

## Due to the error experienced in #5, I applied the binary dummy variable conversion first

bank_bin = pd.get_dummies(bank_df)
bank_bin.head(10)
```

```python
## 6.

## Then applied the min-max normalization

bank_bin_norm = (bank_bin - bank_bin.min()) / (bank_bin.max()-bank_bin.min())
bank_bin_norm.head(10)
```

```python
## 6.

## Output csv

bank_bin_norm.to_csv("bank_numeric.csv", float_format="%1.2f")
```
